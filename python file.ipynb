```python
np.random.seed(2000)
y = np.random.standard_normal((20,2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(7,4)) #size of the image(width and height)
plt.plot(y, lw=1.5)
plt.plot(y, 'k*')
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('My first 2d learnt graph ig')
```

```python
plt.figure(figsize=(7,4)) #size of the image(width and height)
plt.plot(y[:,0], lw=1.5, label='1st')
plt.plot(y[:,1], lw=1.5, label='2nd')
plt.plot(y, 'k*')
plt.grid(True)
plt.axis('tight')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('My first 2d learnt graph ig')
```

```python
y[:,0] = y[:,0]*100
plt.figure(figsize=(7,4)) #size of the image(width and height)
plt.plot(y[:,0], lw=1.5, label='1st')
plt.plot(y[:,1], lw=1.5, label='2st')
plt.plot(y, 'k*')
plt.grid(True)
plt.axis('tight')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('My first 2d learnt graph ig')
```

```python
fig, ax1 = plt.subplots()
plt.plot(y[:,0], lw=1.5, label='1st')
plt.plot(y[:,0], 'k*')
plt.grid(True)
plt.axis('tight')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('My first 2d learnt graph ig')
ax2 =ax1.twinx()
plt.plot(y[:,1], 'g', lw=1.5, label='2nd')
plt.plot(y[:,1], 'k*')
plt.legend(loc=0)
plt.ylabel('Value 2nd')
```
